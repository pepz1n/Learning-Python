# Ordinary Least Squares (OLS) Regression

```python
import statsmodels.api as sm
```

```python
results = sm.OLS(data['votos'], data['pontuação']).fit()
```

```python
results.summary()
```

# Avançado AAA

```python
data[data['ano'] > 1995]
```

```python
data[data['ano'] == 1966]
```

```python
data[(data['ano'] > 1995) & (data['ano'] < 2000)]
```

```python
data[(data['ano'] > 1995) | (data['ano'] < 2000)]
```

```python
data[(data['ano'] > 1995) & (data['ano'] < 2000)].sort_values(by='pontuação', ascending=False).head(10)
```
